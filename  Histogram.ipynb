```python
# Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset and create the 'G_avg' column
df = pd.read_csv('student-por.csv')
df['G_avg'] = (df['G1'] + df['G2'] + df['G3']) / 3
print("Dataset ready.")
```

```python
plt.figure(figsize=(10, 6))
plt.hist(df['age'], bins=8, color='purple', edgecolor='black')
plt.title('Age Distribution of Students', fontsize=16)
plt.xlabel('Age', fontsize=12)
plt.ylabel('Number of Students', fontsize=12)
plt.grid(axis='y', alpha=0.75)

# Save the figure
plt.savefig('8_matplotlib_histogram_age.png', bbox_inches='tight')
plt.show()
```
