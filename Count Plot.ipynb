```python
# Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset and create the 'G_avg' column
df = pd.read_csv('student-por.csv')
df['G_avg'] = (df['G1'] + df['G2'] + df['G3']) / 3
print("Dataset ready.")
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='health', hue='schoolsup', data=df, palette='viridis')
plt.title('Student Health Status by School Support', fontsize=16)
plt.xlabel('Health Status', fontsize=12)
plt.ylabel('Number of Students', fontsize=12)

# Save the figure
plt.savefig('3_seaborn_countplot_health.png', bbox_inches='tight')
plt.show()
```
