```python
# Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset and create the 'G_avg' column
df = pd.read_csv('student-por.csv')
df['G_avg'] = (df['G1'] + df['G2'] + df['G3']) / 3
print("Dataset ready.")
```

```python
internet_counts = df['internet'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(internet_counts, labels=internet_counts.index, autopct='%1.1f%%',
        colors=['lightgreen', 'lightcoral'], startangle=90,
        wedgeprops={'edgecolor': 'black'})
plt.title('Proportion of Students with Internet Access at Home', fontsize=16)

# Save the figure
plt.savefig('7_matplotlib_piechart_internet.png', bbox_inches='tight')
plt.show()
```
