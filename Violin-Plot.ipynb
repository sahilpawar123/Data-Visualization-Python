```python
# Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset and create the 'G_avg' column
df = pd.read_csv('student-por.csv')
df['G_avg'] = (df['G1'] + df['G2'] + df['G3']) / 3
print("Dataset ready.")
```

```python
plt.figure(figsize=(8, 6))
sns.violinplot(x='romantic', y='G_avg', data=df, palette='coolwarm')
plt.title('Final Grades by Romantic Relationship Status', fontsize=16)
plt.xlabel('In a Romantic Relationship ("yes" or "no")', fontsize=12)
plt.ylabel('Average Final Grade', fontsize=12)

# Save the figure
plt.savefig('2_seaborn_violin_romantic.png', bbox_inches='tight')
plt.show()
```
