```python
# Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset and create the 'G_avg' column
df = pd.read_csv('student-por.csv')
df['G_avg'] = (df['G1'] + df['G2'] + df['G3']) / 3
print("Dataset ready.")
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['failures'], df['G_avg'], alpha=0.5, color='red')
plt.title('Final Grade vs. Number of Past Failures', fontsize=16)
plt.xlabel('Number of Past Class Failures', fontsize=12)
plt.ylabel('Average Final Grade', fontsize=12)
plt.xticks([0, 1, 2, 3])
plt.grid(True)

# Save the figure
plt.savefig('9_matplotlib_scatterplot_failures.png', bbox_inches='tight')
plt.show()
```
