```python
# Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset and create the 'G_avg' column
df = pd.read_csv('student-por.csv')
df['G_avg'] = (df['G1'] + df['G2'] + df['G3']) / 3
print("Dataset ready.")
```

```python
g = sns.pairplot(df[['G_avg', 'studytime', 'traveltime', 'absences']], hue='studytime', palette='rocket')
g.fig.suptitle('Pair Plot of Grades, Study Time, Travel Time, and Absences', y=1.02)

# Save the figure
plt.savefig('5_seaborn_pairplot_overview.png', bbox_inches='tight')
plt.show()
```
