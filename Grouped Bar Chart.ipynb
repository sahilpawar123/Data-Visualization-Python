```python
# Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset and create the 'G_avg' column
df = pd.read_csv('student-por.csv')
df['G_avg'] = (df['G1'] + df['G2'] + df['G3']) / 3
print("Dataset ready.")
```

```python
paid_study_time = df.groupby('paid')['studytime'].value_counts(normalize=True).unstack() * 100
ax = paid_study_time.plot(kind='bar', figsize=(12, 7), colormap='plasma')
plt.title('Study Time Distribution by Paid Class Status', fontsize=16)
plt.xlabel('Took Extra Paid Classes', fontsize=12)
plt.ylabel('Percentage of Students (%)', fontsize=12)
plt.xticks(rotation=0)
plt.legend(title='Study Time', labels=['< 2 hrs', '2-5 hrs', '5-10 hrs', '> 10 hrs'])

# Save the figure
plt.savefig('10_matplotlib_grouped_bar_studytime.png', bbox_inches='tight')
plt.show()
```
