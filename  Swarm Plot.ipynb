```python
# Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset and create the 'G_avg' column
df = pd.read_csv('student-por.csv')
df['G_avg'] = (df['G1'] + df['G2'] + df['G3']) / 3
print("Dataset ready.")
```

```python
plt.figure(figsize=(12, 7))
sns.swarmplot(x='Walc', y='G_avg', data=df, palette='magma', s=4)
plt.title('Final Grades vs. Weekend Alcohol Consumption', fontsize=16)
plt.xlabel('Weekend Alcohol Consumption Level', fontsize=12)
plt.ylabel('Average Final Grade', fontsize=12)

# Save the figure
plt.savefig('4_seaborn_swarmplot_alcohol.png', bbox_inches='tight')
plt.show()
```
