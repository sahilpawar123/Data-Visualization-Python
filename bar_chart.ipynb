```python
# Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset and create the 'G_avg' column
df = pd.read_csv('student-por.csv')
df['G_avg'] = (df['G1'] + df['G2'] + df['G3']) / 3
print("Dataset ready.")
```

```python
plt.figure(figsize=(8, 6))
# The fix: add hue='famsup' and legend=False
sns.boxplot(x='famsup', y='G_avg', data=df, hue='famsup', palette='pastel', legend=False)
plt.title('Final Grades by Family Educational Support', fontsize=16)
plt.xlabel('Family Support ("yes" or "no")', fontsize=12)
plt.ylabel('Average Final Grade', fontsize=12)

# Save the figure
plt.savefig('1_seaborn_boxplot_famsup.png', bbox_inches='tight')
plt.show()
```
