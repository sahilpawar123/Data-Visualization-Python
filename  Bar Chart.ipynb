```python
# Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset and create the 'G_avg' column
df = pd.read_csv('student-por.csv')
df['G_avg'] = (df['G1'] + df['G2'] + df['G3']) / 3
print("Dataset ready.")
```

```python
parent_status_grades = df.groupby('Pstatus')['G_avg'].mean()
plt.figure(figsize=(7, 6))
plt.bar(parent_status_grades.index, parent_status_grades.values, color=['skyblue', 'salmon'])
plt.title("Average Final Grade by Parents' Cohabitation Status", fontsize=16)
plt.xlabel("Parent's Status ('T' = Together, 'A' = Apart)", fontsize=12)
plt.ylabel("Average Final Grade", fontsize=12)
plt.ylim(10, 12)

# Save the figure
plt.savefig('6_matplotlib_barchart_pstatus.png', bbox_inches='tight')
plt.show()
```
